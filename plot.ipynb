```python
from utils import *
import sys
import os
wd = os.getcwd()
sys.path.append(wd)
```

```python
with open('RNN.pkl', 'rb') as file:
    RNN = pickle.load(file)
with open('iteration.pkl', 'rb') as file:
    DMFT = pickle.load(file)[-1]

t = 10.
Delta_t = .1
x = np.arange(0, t, Delta_t)
```

```python
R_RNN = np.array([np.mean(np.array([RNN[-1][i+k,i] for i in range(RNN[-1].shape[0]-k)])) for k in range(RNN[-1].shape[0])])
R_DMFT = np.array([np.mean(np.array([DMFT[1][i+k,i] for i in range(DMFT[1].shape[0]-k)])) for k in range(DMFT[1].shape[0])])
```

```python
plt.plot(x, RNN[1], label='$m(t)$')
plt.plot(x, np.diag(RNN[3]), label='$C(t,t)$')
plt.plot(x, R_RNN, label='$R(t-t^\prime)$')

plt.plot(x, DMFT[3], ls='--', c='k', label='DMFT')
plt.plot(x, np.diag(DMFT[0]), ls='--', c='k')
plt.plot(x, R_DMFT, ls='--', c='k')

plt.xlabel('$t/t-t^\prime$')
plt.ylabel('observables')
plt.legend()
plt.savefig(wd+'/comparison.pdf',bbox_inches='tight')
```
